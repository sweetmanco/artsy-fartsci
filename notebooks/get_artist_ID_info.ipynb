```python
import pandas as pd

dct = {
    'name': [],
    'id': [],
    'gender': [],
    'birthday': [],
    'deathday': [],
    'hometown': [],
    'location': [],
    'nationality': [],
    'sortable_name': [],
    'slug': []
}

for x in list_:
    
    embedded = x.get('_embedded', {})
    artists = embedded.get('artists', [])

    if not artists:
        continue  
    info = artists[0]
    info = x['_embedded']['artists'][0]
    name = info.get('name', None)
    id = info.get('id', None)
    gender = info.get('gender', None)
    birthday = info.get('birthday', None)
    deathday = info.get('deathday', None)
    hometown = info.get('hometown', None)
    location = info.get('location', None)
    nationality = info.get('nationality', None)
    sortable_name = info.get('sortable_name', None)
    slug = info.get('slug', None)

    dct['name'].append(name)
    dct['id'].append(id)
    dct['gender'].append(gender)
    dct['birthday'].append(birthday)
    dct['deathday'].append(deathday)
    dct['hometown'].append(hometown)
    dct['location'].append(location)
    dct['nationality'].append(nationality)
    dct['sortable_name'].append(sortable_name)
    dct['slug'].append(slug)

df = pd.DataFrame(dct)
```
